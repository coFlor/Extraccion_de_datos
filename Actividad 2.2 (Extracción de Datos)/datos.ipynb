```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('Datos de Facturación.xlsx')
df.info()
```

```python
#ACÁ FILTRAMOS CVE_CLPV DE 1000 A 2000 usando pd.to_numeric
df["CVE_CLPV"] = pd.to_numeric(df["CVE_CLPV"], errors='coerce')
filtro1 = df[(df["CVE_CLPV"] > 999) & (df["CVE_CLPV"] < 2001)]
filtro1
```

```python
#acá hicimos que se muestre todas las clases menos la 5 y 4 de CVE_VEND
filtro2=df[(~(df["CVE_VEND"] == "5")) & (~(df["CVE_VEND"] == "4"))]
filtro2
```

```python
#correspondiente a la fecha 28/02/2022 se filtran los datos
df["FECHA_ENT"] = pd.to_datetime(df["FECHA_ENT"],errors='coerce')
filtro3=df[(df["FECHA_ENT"] == "28/02/2022")]
filtro3
```

```python
#se muestran los datos menores a 5951.7 de "CAN_NOT" O tambien con el filtro de E
filtro4=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
```

```python
#acá sólo se hace aparicion de las columnas CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro5=df.iloc[:,[0,6,7,9]]
filtro5
```

```python
#Sólo se muestran las filas 70001 y 7100

filtro6= df.iloc[7001:7100,:]
filtro6
```
