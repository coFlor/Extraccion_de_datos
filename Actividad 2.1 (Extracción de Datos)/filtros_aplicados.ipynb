```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df.info()
```

```python
#Filtro por objeto a los clientes que en este caso es el IPN
filtro1 = df[df["NOMBRE_CLIENTE"] == "INSTITUTO POLITECNICO NACIONAL"]
filtro1
```

```python
filtro1.to_csv("FILTRO1.csv")
#df=
```

```python
#Filtro por filas, iniciando de la posicion 24, y va de 100 en 100
filtro2= df.iloc[24::100,:]
filtro2
```

```python
filtro2.to_csv("FILTRO2.csv")
#df=
```

```python
#Filtro por columnas a los datos de: nombre cliente y CANT
#filtro3=df.iloc[ : , 2:5]
filtro3= df.iloc[ : , 2:5]
filtro3
```

```python
filtro3.to_csv("FILTRO3.csv")
#df=
```

```python
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python
#Aplicamos filtro de filas vs columnas. Donde muestro la cantidad de ventas de Aaron
filtro4=df1.loc[["AARON NUÑO CORDOVA"], ["CANT"]]
filtro4
```

```python
filtro4.to_csv("FILTRO4.csv")
#df=
```

```python
#Filtro por cabecera se filtran los primeros 5 datos, iniciando desde el 0
filtro5=df.head()
filtro5
```

```python
filtro5.to_csv("FILTRO5.csv")
#df=
```

```python
#Filtro por cola se muestrran los ultimos 5 datos de toda la base.
filtro6=df.tail()
filtro6
```

```python
filtro6.to_csv("FILTRO6.csv")
#df=
```

```python
#filtro por comparacion comparo a que CANT sea mayor a 30 
filtro7=df[df["CANT"] > 30]
filtro7
```

```python
filtro7.to_csv("FILTRO7.csv")
#df=
```

```python
#Filtro Y, muestro que CANT sea menor que 30 Y que sea en la fecha 2022-03-14
filtro8 = df[(df["CANT"] < 30) & (df["FECHA_DOC"] == pd.Timestamp("2022-03-14"))]
filtro8
```

```python
filtro8.to_csv("FILTRO8.csv")
#df=
```

```python
#Filtro O, muestro que CANT sea menor que 30 O que sea en la fecha 2022-12-23
filtro9 = df[(df["CANT"] < 30) | (df["FECHA_DOC"] == pd.Timestamp("2022-12-23"))]
filtro9
```

```python
filtro9.to_csv("FILTRO9.csv")
#df=
```

```python
#Filtro not. Hago que NO se muestren los datos con la fecha 2022-12-23
filtro10 = df[~(df["FECHA_DOC"] == pd.Timestamp("2022-12-23"))]
filtro10
```

```python
filtro10.to_csv("FILTRO10.csv")
#df=
```
